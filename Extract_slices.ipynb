```python
from matplotlib import cm, pyplot as plt
import numpy as np
import os
import fdsreader as fds

sim = fds.Simulation("/Users/jaroslavcepl/Documents/Projekty/logo_noze/Ohen_Final/")
#print(sim.slices)
slc = sim.slices[1]
data, coordinates = slc.to_global(masked=True, fill=np.nan, return_coordinates=True)
#print(coordinates)
    # Get the first slice
#    slc = sim.slices[1]
 #   data, coordinates = slc.to_global(masked=True, fill=np.nan, return_coordinates=True)

    # Set colormap
cmap = cm.get_cmap('coolwarm').copy()
#print(cmap)
    # Set obsts color
cmap.set_bad('white')
#print(data)

output_dir = "/Users/jaroslavcepl/Documents/Projekty/logo_noze/Ohen_Final/SlicesOutput/"
os.makedirs(output_dir, exist_ok=True)

    # Plot the slice
for i in range(0, 30):
  plt.imshow(data[i].T, vmin=0, vmax=750, origin="lower", cmap=cmap)
  plt.colorbar()
  plt.show()
  slice_data = data[i].T  # Transpose for correct orientation
  output_file = os.path.join(output_dir, f"slice_{i:03d}.txt")
  np.savetxt(output_file, slice_data, delimiter=",", fmt="%.3f")
  print(f"Saved slice {i} to {output_file}")
    

#if __name__ == "__main__":
#    main()
```
